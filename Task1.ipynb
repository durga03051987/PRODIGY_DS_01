### Distrubution of population in a specific year over categorical variable i.e country names, for categorical variable consider the barchart

```python
#Here consider the year 2022 for caluculating the population over countries
```

```python
df_2022 =pop_data[['Country Name','2022']]
```

```python
df_2022
```

```python
# Consider the distibution of population for top 10 countries
```

```python
top10=df_2022.sort_values('2022',ascending=False).head(10)
top10
```

```python
# Distribution of population for top 10 countries with categorical variable Country Name 
```

### Let us  consider  ðŸ“Š Bar Chart for population of top 10 countries in 2022

```python
plt.figure(figsize=(12,6))
plt.bar(top10['Country Name'],top10['2022'],color='green')
plt.xticks(rotation=45)
plt.xlabel("Country Names")
plt.ylabel("Population in Billions")
plt.title("TOP 10 Most populous countries in 2022")
plt.show()
```

```python
## Summary  :  TOP 10 Most populous Countries  in 2022
```
